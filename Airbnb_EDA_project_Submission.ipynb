#### Chart - 6

 **Price Range Vs Affordability**

 Compare the price range with Affordability parameter

````python
# Chart - 6 visualization code
# defineing the price categories
def categorize_price(price):
  if price <= 150:
    return 'Budget-friendly'
  elif 150 < price <= 500:
    return 'Affordable'
  else:
    return 'Expensive'
````

````python
data_df['price_category'] = data_df['price'].apply(categorize_price)
#ploting a count plot
plt.figure(figsize=(10,6))
sns.countplot(x= 'price_category', data= data_df, palette= 'viridis')
plt.title('Count of Listing by Price Category')
plt.xlabel('Price Category')
plt.ylabel('Count')
plt.show()
````

Host Verification:

 What percentage of hosts have been verified? Provide a breakdown of verified and non-verified hosts

````python
total_hosts = len(data_df['host_id'].unique()) #total number of hosts
verified_hosts =len(data_df[data_df['host_id'].notna()]['host_id'].unique()) #number of varified host
non_verified_hosts = total_hosts - verified_hosts #number of non verified host
percentage_verified = (verified_hosts / total_hosts) *100 #percentage of verified hosts

#creating a bar plot
plt.figure(figsize=(8,6))
sns.barplot(x=['Verified hosts','Non verified hosts'], y= [verified_hosts, non_verified_hosts])
plt.title('host verification brakedown')
plt.ylabel('number of host')
plt.xlabel('verification status')
plt.text(0, verified_hosts + 50 , f'{percentage_verified:.2f}%', ha= 'center', fontsize=10) #logic from google
plt.ylim(0, total_hosts + 100)
plt.show()

print(f'Percentage of veerified hosts: {percentage_verified:.2f}%')
````

#### Chart - 11 - Correlation Heatmap

````python
# Correlation Heatmap visualization code
````

````python
#selecting numerical coloumn for correlation
numerical_columns = data_df.select_dtypes(include=['int64','float64'])
#calculate correlation matrix
correlation_matrix = numerical_columns.corr()

#ploting a correlation plot
plt.figure(figsize= (10,8))
sns.heatmap(correlation_matrix, annot= True, cmap= 'coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap for Numerical Variables')
plt.show()
````

Answer Here.

##### 2. What is/are the insight(s) found from the chart?

Positive Correlation: If two variables have a positive correlation (values closer to 1), it means that as one variable increases, the other tends to increase as well. For example, if you observe a positive correlation between 'avaibility_365' and 'price' it suggests that listings with more availabilty per month tend to have a higher price.

Negative Correlation: A negative correlation (values closer to -1) indicates that as one variable increases, the other tends to decrease.

No Correlation: Values close to 0 indicate little to no linear correlation between variables.

Multicollinearity: High positive or negative correlations between pairs of independent variables (features) can indicate multicollinearity, which can affect the performance of predictive models.
